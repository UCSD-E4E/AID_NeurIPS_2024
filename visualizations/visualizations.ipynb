```python
def get_data(file):
    df = pd.read_csv(file)
    df["ANNO COUNT"] = df["TRUE POSITIVE"] + df["FALSE POSITIVE"]
    df = df[df["ANNO COUNT"] > 30]
    df = df.sort_values("F1", ascending=True)
    display(df.head(5)) 
    display(df.tail(5))
    return df

fg = get_data("fg_bg_class_stats.csv")
fg
```

```python
#!pip install altair
import altair as alt

fg["type"] = "Foreground background seperator"
tn["type"] = "Tweetynet"
mf["type"] = "Mircofaune"
df = pd.concat([fg, tn, mf])
df

alt.Chart(df).mark_bar().encode(
    alt.X(alt.repeat("row"), bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet("type", title="")
).properties(
    width=150,
    height=150
).repeat(
    row=["PRECISION", "RECALL", "F1", "ANNO COUNT"]
).properties(
    title="Distrubution of Scores Accross Species"
)
```

```python
one = alt.Chart(df).mark_bar().encode(
    alt.X("PRECISION", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet("type", title=None, header=None),
    color="type"
).properties(
    width=150,
    height=150
).properties(
    title="Precision"
)


two = alt.Chart(df).mark_bar().encode(
    alt.X("RECALL", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet("type", title=""),
    color="type"
).properties(
    width=150,
    height=150
).properties(
    title="Recall"
)


three = alt.Chart(df).mark_bar().encode(
    alt.X("F1", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet("type", title=""),
    color="type"
).properties(
    width=150,
    height=150
).properties(
    title="F1"
)


four = alt.Chart(df).mark_bar().encode(
    alt.X("ANNO COUNT", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet("type", title=""),
    color="type"
).properties(
    width=150,
    height=150
).properties(
    title="Annotation Counts"
)


(one & two & three & four).properties(
    title="Performance of Models Across Species"
)

one
```

```python
fg["type"] = "FG_BG_SEP"
tn["type"] = "TweetyNet"
mf["type"] = "Microfaune"
df = pd.concat([fg, tn, mf])
df


width = 90
height = 80

one = alt.Chart(df).mark_bar().encode(
    alt.X("PRECISION", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet(
        "type", 
        title=None, 
        header=alt.Header(title="Precision",labelFontSize=0, labelColor="white", titleFont='Times')
    ),
    color="type"
).properties(
    width=width,
    height=height
)

two = alt.Chart(df).mark_bar().encode(
    alt.X("RECALL", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet(
        "type", 
        title=None, 
        header=alt.Header(title="Recall",labelFontSize=0, labelColor="white", titleFont='Times')
    ),
    color="type"
).properties(
    width=width,
    height=height
)

three = alt.Chart(df).mark_bar().encode(
    alt.X("F1", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet(
        "type", 
        title=None, 
        header=alt.Header(title="F1",labelFontSize=0, labelColor="white", titleFont='Times')
    ),
    color="type"
).properties(
    width=width,
    height=height
)

four = alt.Chart(df).mark_bar().encode(
    alt.X("ANNO COUNT", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet(
        "type", 
        title=None, 
        header=alt.Header(title="Annotation Counts",labelFontSize=0, labelColor="white", titleFont='Times')
    ),
    color=alt.Color("type", legend=alt.Legend(orient="top"))
).properties(
    width=width,
    height=height
)


(one & two & three & four).properties(
    title=alt.Title("Performance of Models Across Species",
                    font='Times')
).configure_axis(
    titleFontSize=10,
    titleFontWeight='normal'
)
```

```python
df['type'].unique()
fgbg = df[df['type'] == 'FG_BG_SEP']
twt = df[df['type'] == 'TweetyNet']
mf = df[df['type'] == 'Microfaune']
```

```python
plt.rcParams.update({'font.family': 'serif', 'font.serif': 'Times'})
fig = plt.figure(figsize=(10, 7), constrained_layout=True)
subfigs = fig.subfigures(nrows=3, ncols=1, wspace=0.07)
axs = [subfig.subplots(nrows=1, ncols=3, 
                       gridspec_kw={'wspace': 0},
                       sharey=True) 
       for subfig in subfigs.ravel()]

####################################################################################
subfigs[0].suptitle('Precision', fontsize=20, fontweight='bold')
axs[0][0].hist(fgbg['PRECISION'])
axs[0][0].set(title="FG BG Separation")
axs[0][0].set_ylabel('Count')
axs[0][0].set_xlabel('Precision')

axs[0][1].hist(twt['PRECISION'], color='tab:orange')
axs[0][1].set(title="Tweetynet")
axs[0][1].set_xlabel('Precision')

axs[0][2].hist(mf['PRECISION'], color='tab:red')
axs[0][2].set(title="Microfaune")
axs[0][2].set_xlabel('Precision')

####################################################################################
subfigs[1].suptitle('Recall', fontsize=20, fontweight='bold')
axs[1][0].hist(fgbg['RECALL'])
axs[1][0].set(title="FG BG Separation")
axs[1][0].set_ylabel('Count')
axs[1][0].set_xlabel('Recall')

axs[1][1].hist(twt['RECALL'], color='tab:orange')
axs[1][1].set(title="Tweetynet")
axs[1][1].set_xlabel('Recall')

axs[1][2].hist(mf['RECALL'], color='tab:red')
axs[1][2].set(title="Microfaune")
axs[1][2].set_xlabel('Recall')

####################################################################################
subfigs[2].suptitle('F1', fontsize=20, fontweight='bold')
axs[2][0].hist(fgbg['F1'])
axs[2][0].set(title="FG BG Separation")
axs[2][0].set_ylabel('Count')
axs[2][0].set_xlabel('F1')

axs[2][1].hist(twt['F1'], color='tab:orange')
axs[2][1].set(title="Tweetynet")
axs[2][1].set_xlabel('F1')

axs[2][2].hist(mf['F1'], color='tab:red')
axs[2][2].set(title="Microfaune")
axs[2][2].set_xlabel('F1')

####################################################################################
# subfigs[3].suptitle('Annotation Counts', fontsize=20, fontweight='bold')
# axs[3][0].hist(fgbg['ANNO COUNT'])
# axs[3][0].set(title="FG BG Separation")

# axs[3][1].hist(twt['ANNO COUNT'], color='tab:orange')
# axs[3][1].set(title="Tweetynet")

# axs[3][2].hist(mf['ANNO COUNT'], color='tab:red')
# axs[3][2].set(title="Microfaune")

fig.suptitle('Performance of Models Across Species', fontsize=25, fontweight='extra bold')
```
