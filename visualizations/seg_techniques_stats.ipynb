```python
data, rate = librosa.load("XC36415 - Crested Eagle - Morphnus guianensis.mp3")
```

```python
spec = librosa.feature.melspectrogram(y=data[:rate * 3], power=0.25)
plt.imshow(spec)
```

```python
librosa
```

```python
plt.rcParams.update({'font.family': 'serif', 'font.serif': 'Times'})
ratio, spec = perform_stft(data[:rate*3], SAMPLE_RATE=rate)
print(ratio)
spec.shape
spec = spec[:120, :]
```

```python

fig, ax = plt.subplots(5, figsize=(10,15))
librosa.display.specshow(spec ** .25, 
                         ax=ax[0],
                         y_axis='linear',
                         x_axis='s',
                         sr=rate,
                         hop_length=ratio,
                         cmap='hot')
ax[0].set(title='Crested Eagle Call from XC36415')

time_med, freq_med = calculate_medians(spec)
bin_spec = binary_thresholding(spec, time_med, freq_med)
librosa.display.specshow(bin_spec, 
                         ax=ax[1],
                         y_axis='linear',
                         x_axis='s',
                         sr=rate,
                         hop_length=ratio,
                         cmap='hot')
ax[1].set(title='Binary Mask')

bin_opened = binary_morph_opening(bin_spec)
librosa.display.specshow(bin_opened,
                         ax=ax[2],
                         y_axis='linear',
                         x_axis='s',
                         sr=rate,
                         hop_length=ratio,
                         cmap='hot')
ax[2].set(title='Binary Mask after Morph Opening')

temp_vector = temporal_thresholding(bin_opened)
ax[3].scatter(x=np.arange(len(temp_vector)) * ratio / rate, 
              y=temp_vector,
              sizes=[3]*len(temp_vector))
ax[3].set(title='Temporal Indicator Vector',
          xlabel='Time (seconds)',
          ylabel='Value')
ax[3].margins(x=0)
ax[3].set_xticks(ticks=[0, 0.5, 1, 1.5, 2, 2.5])
ax[3].grid(axis='x')

dilated_tiv = indicator_vector_processing(temp_vector)[0]
ax[4].scatter(x=np.arange(len(dilated_tiv)) * ratio / rate, 
              y=dilated_tiv,
              sizes=[3]*len(dilated_tiv))
ax[4].set(title='Dilated Temporal Indicator Vector',
          xlabel='Time (seconds)',
          ylabel='Value')
ax[4].margins(x=0)
ax[4].set_xticks(ticks=[0, 0.5, 1, 1.5, 2, 2.5])
ax[4].grid(axis='x')

plt.tight_layout()
```

```python
fg_bg_pre = pd.read_csv("segmentation_csvs/fg_bg_pre.csv")
fg_bg_pre["TYPE"]  = "fg_bg_pre"
manual = pd.read_csv("segmentation_csvs/manual_pre.csv")
manual["TYPE"]  = "manual"
microfaune = pd.read_csv("segmentation_csvs/microfaune_pre.csv")
microfaune["TYPE"]  = "microfaune"
tweetynet = pd.read_csv("segmentation_csvs/tweetynet_pre.csv")
tweetynet["TYPE"]  = "tweetynet"

df = pd.concat([microfaune, fg_bg_pre, tweetynet, manual])
df.head()
```

```python
import numpy as np
df["log_duration"] = df["DURATION"].apply(np.log).to_list()
df = df.reset_index()
ax = sns.boxplot(data=df, x="TYPE", y="log_duration")
ax.set_yticks(np.arange(-3, 4))
ax.set_yticklabels(10.0**np.arange(-3, 4))
```

```python
plt.rc('font', family='serif', serif='Times')
plt.rc('text', usetex=True)
plt.rc('xtick', labelsize=8)
plt.rc('ytick', labelsize=8)
plt.rc('axes', labelsize=8)
import seaborn as sns
sns.set_theme(font_scale=1.5, font='Times New Roman')

names = {
    'microfaune': 'Microfaune',
    'fg_bg_pre': 'FG BG Separation',
    'tweetynet': 'Tweetynet',
    'manual': 'Manual Labeling'
}

fig, ax = plt.subplots(figsize=(10, 8))

ax = sns.boxenplot(data=df.assign(TYPE=df['TYPE'].apply(lambda x: names[x])), 
                   x="TYPE", y="log_duration", width=0.6, color='steelblue');
ax.set_yticks(np.arange(-5, 6))
ax.set_yticklabels(10.0**np.arange(-5, 6))
ax.set_ylabel("Log Duration")
ax.set_xlabel("Method")
ax.set_title("Log Distribution of Annotation Durations")
plt.tight_layout()
```
