```python
def get_data(file):
    df = pd.read_csv(file)
    df["ANNO COUNT"] = df["TRUE POSITIVE"] + df["FALSE POSITIVE"]
    df = df[df["ANNO COUNT"] > 30]
    df = df.sort_values("F1", ascending=True)
    display(df.head(5)) 
    display(df.tail(5))
    return df

fg = get_data("fg_bg_class_stats.csv")
fg
```

```python
#!pip install altair
import altair as alt

fg["type"] = "Foreground background seperator"
tn["type"] = "Tweetynet"
mf["type"] = "Mircofaune"
df = pd.concat([fg, tn, mf])
df

alt.Chart(df).mark_bar().encode(
    alt.X(alt.repeat("row"), bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet("type", title="")
).properties(
    width=150,
    height=150
).repeat(
    row=["PRECISION", "RECALL", "F1", "ANNO COUNT"]
).properties(
    title="Distrubution of Scores Accross Species"
)
```

```python
one = alt.Chart(df).mark_bar().encode(
    alt.X("PRECISION", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet("type", title=None, header=None),
    color="type"
).properties(
    width=150,
    height=150
).properties(
    title="Precision"
)


two = alt.Chart(df).mark_bar().encode(
    alt.X("RECALL", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet("type", title=""),
    color="type"
).properties(
    width=150,
    height=150
).properties(
    title="Recall"
)


three = alt.Chart(df).mark_bar().encode(
    alt.X("F1", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet("type", title=""),
    color="type"
).properties(
    width=150,
    height=150
).properties(
    title="F1"
)


four = alt.Chart(df).mark_bar().encode(
    alt.X("ANNO COUNT", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet("type", title=""),
    color="type"
).properties(
    width=150,
    height=150
).properties(
    title="Annotation Counts"
)


(one & two & three & four).properties(
    title="Performance of Models Across Species"
)

one
```

```python


fg["type"] = "FG_BG_SEP"
tn["type"] = "TweetyNet"
mf["type"] = "Microfaune"
df = pd.concat([fg, tn, mf])
df


width = 90
height = 80

one = alt.Chart(df).mark_bar().encode(
    alt.X("PRECISION", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet(
        "type", 
        title=None, 
        header=alt.Header(title="Precision",labelFontSize=0, labelColor="white")
    ),
    color="type"
).properties(
    width=width,
    height=height
)

two = alt.Chart(df).mark_bar().encode(
    alt.X("RECALL", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet(
        "type", 
        title=None, 
        header=alt.Header(title="Recall",labelFontSize=0, labelColor="white")
    ),
    color="type"
).properties(
    width=width,
    height=height
)

three = alt.Chart(df).mark_bar().encode(
    alt.X("F1", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet(
        "type", 
        title=None, 
        header=alt.Header(title="F1",labelFontSize=0, labelColor="white")
    ),
    color="type"
).properties(
    width=width,
    height=height
)

four = alt.Chart(df).mark_bar().encode(
    alt.X("ANNO COUNT", bin=True, title=""),
    y=alt.Y("count()", title="number of species"),
    facet=alt.Facet(
        "type", 
        title=None, 
        header=alt.Header(title="Annotation Counts",labelFontSize=0, labelColor="white")
    ),
    color=alt.Color("type", legend=alt.Legend(orient="top"))
).properties(
    width=width,
    height=height
)


(one & two & three & four).properties(
    title="Performance of Models Across Species"
).configure_axis(
    titleFontSize=10,
    titleFontWeight='normal'
)
```
